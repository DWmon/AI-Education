# Data load

```python
import numpy as np
import pandas as pd
news_data = pd.read_csv('./data/news.csv')
```

```python
news_data
```

# Preprocessing

```python
import re

def preprocessing(sentence):
    sentence =re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z]', ' ', sentence)
    return sentence
```

```python
news_data['content_cleaned'] = news_data['content'].apply(preprocessing)
content = news_data['content_cleaned'].tolist()
```

```python
content[0]
```

# Kmeans Clustering

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
```

```python
np.unique(news_data.category)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(content)
```

```python
X = normalize(X)
```

```python
km = KMeans(3)
km.fit(X)
```

# Result

```python
news_data['labels'] = km.labels_
```

```python
news_data[['category','content','labels']]
```
